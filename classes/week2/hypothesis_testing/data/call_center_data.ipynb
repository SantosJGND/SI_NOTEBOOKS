### Call Center data set

This data set is from a call center. The data set contains information about the calls made by the call center. The original version can be found on kaggle [here](https://www.kaggle.com/datasets/ashishpandey5210/call-center-dataset).

The data set presented here and to be used in class is a modified version of the original data set. The modifications are:

- A `datetime` column was created by merging the `date` and `time` columns.
- An `interval` column was created by calculating the time difference between successive calls.
- The data set was subset to only include the two columns `datetime` and `interval`.

### Importing the data set

```python
import pandas as pd

call_center_df= pd.read_csv("data/call_center_small.csv")
call_center_df.head()
```

## Visualizing the data set

This data set was selected in order to study arrival times. We can inspect whether this data conforms to the exponential distribution by plotting the histogram of the data set.


```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.hist(call_center_df[call_center_df["interval"] < 20000]["interval"], bins= 20)
plt.xlabel("Inter-arrival time (minutes)")
plt.ylabel("Frequency")
plt.title("Distribution of inter-call times")
plt.show()
```
