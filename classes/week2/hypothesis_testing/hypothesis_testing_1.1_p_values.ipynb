## P-value 

### 1. P-value. 

**Recap** Remember how we constructed confidence intervals: we began by studying the cdf of the sample mean, and then we found the interval that contained 95% of the probability.We first looked at the CDF of the lower bound, to the left of the mean. 

In the context of the standard normal distribution, we saw that a value of -1.96 corresponded to the 2.5th percentile.

The idea of this section is that, in the context of the standard normal, we can say that a value of -1.96 corresponds to a p-value of 0.025.
  - i.e. 2.5% of the probability is to the left of -1.96.

**Definition** The p-value is the probability of observing a value as extreme or more extreme than the observed value, given that the null hypothesis is true .

This means that we always look at the tail of the distribution that corresponds to the alternative hypothesis. if the test statistic is to the left of the mean we take the cdf, if it is to the right of the mean we take the *survival function*, `1-cdf`.

This answers the question : "How likely is it that we would observe a value as extreme or more extreme than the observed value, given that the null hypothesis is true?"

### Example

We own a restaurant and we "know" that the average tip is 15%. We assume that the tips are normally distributed with a standard deviation of 5%. How likely is it that we would get a tip of 25% or more?

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Set the parameters for the null hypothesis
null_mean = 15
null_std = 5

# Set the observed value
observed_tip = 25

# Generate the x values for the plot
x = np.linspace(null_mean - 3 * null_std, null_mean + 3 * null_std, 100)

# Generate the y values for the null hypothesis distribution
y_null = stats.norm.pdf(x, null_mean, null_std)

# Calculate the p-value under the null hypothesis
z_score = (observed_tip - null_mean) / null_std
p_value = 1 - stats.norm.cdf(z_score)

# Create the plot
fig, ax = plt.subplots()
ax.plot(x, y_null, 'b-', linewidth=2)
ax.axvline(observed_tip, color='r', linestyle='--')
ax.annotate(f'p-value = {p_value:.4f}', xy=(observed_tip, 0.02), xytext=(10, 0.02),
            arrowprops=dict(facecolor='black', arrowstyle='->'))
ax.set_xlabel('Tip Percentage')
ax.set_ylabel('Probability Density')
ax.set_title('Null Hypothesis Distribution')
plt.show()
# # 


```

##  Summary

- The p-value is the probability of observing a value as extreme or more extreme than the observed value, given that the null hypothesis is true.

- The p-value is always taken from your **Null Distribution**, i.e. the smaller it is, the more likely it is that the null hypothesis is false and the alternative hypothesis is true.
