  > But we still assumed the nature of the distribution. 


#### Kolmogorov-Smirnov test


The Kolmogorov-Smirnov testis used to test whether two samples are drawn from the same distribution. It compares the cumulative distribution functions of two samples.

The test is nonparametric because the distribution of the test statistic does not depend on the form of the underlying cumulative distribution function. 



```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate data for the normal distribution
x_normal = np.linspace(-4, 4, 100)
y_normal = stats.norm.cdf(x_normal, 0, 1)

# Generate data for the exponential distribution
x_exponential = np.linspace(0, 4, 100)
y_exponential = stats.expon.cdf(x_exponential, 0, 1)

diff= len(x_normal) - len(x_exponential)
y_exp_for_plotting = [0] * diff + list(y_exponential)
# Plot the CDF of the normal distribution
plt.plot(x_normal, y_normal, label='Normal Distribution')

# Plot the CDF of the exponential distribution
plt.plot(x_exponential, y_exponential, label='Exponential Distribution')

# Set plot title and labels
plt.title('CDF of Normal and Exponential Distributions')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

#### Kolmogorov-Smirnov test in Python 

```python
import numpy as np
import scipy.stats as stats

# Set random seed for reproducibility
np.random.seed(0)

# Generate a sample from the uniform distribution
sample_uniform = np.random.uniform(0, 1, 100)

# Generate a sample from the normal distribution
sample_normal = np.random.normal(0, 1, 100)

# Perform the Kolmogorov-Smirnov test for the uniform distribution
ks_statistic_uniform, p_value_uniform = stats.kstest(sample_uniform, 'uniform')

# Perform the Kolmogorov-Smirnov test for the normal distribution
ks_statistic_normal, p_value_normal = stats.kstest(sample_normal, 'norm')

# Print the results
print(f"Kolmogorov-Smirnov test for uniform distribution:")
print(f"KS statistic: {ks_statistic_uniform}")
print(f"P-value: {p_value_uniform}")
print()
print(f"Kolmogorov-Smirnov test for normal distribution:")
print(f"KS statistic: {ks_statistic_normal}")
print(f"P-value: {p_value_normal}")
```
