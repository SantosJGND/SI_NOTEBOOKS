### Manipulating Distributions.

In this notbebook, we will resort to a python library called [scipy](https://docs.scipy.org/doc/scipy/reference/stats.html) to manipulate distributions. 

**Reminder** : The `scipy.stats` module contains a large number of probability distributions as well as a growing library of statistical functions. Each univariate distribution is an instance of a subclass of `rv_continuous` (for continuous distributions) or of `rv_discrete` (for discrete distributions). Both share a common set of methods.

In previous notebooks, we studied the methods `pdf`, `pmf`, `cdf` and `ppf` to study distributions. In this notebook we are looking at the `rvs` method. The `rvs` method generates random numbers from a distribution.


Let's start with the Bonomial distribution.


### Example: Binomial Distribution

```python
## simulate variables using the binomial distribution with p= 0.5 and 1000 trials

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# set the seed
np.random.seed(123)

# distriubtion parameters
p = 0.5
n = 1000

# distribution
binomial = stats.binom(n, p)

# simulate 1000 draws from the distribution
simulated = binomial.rvs(1000)

print(f"the first 10 draws are: {simulated[:10]}")

print("mean: ", np.mean(simulated))
print("variance: ", np.var(simulated))
print("standard deviation: ", np.std(simulated))
```

Notice how the average is very close to that expected from theory: $np = 1000 \times 0.5 = 500$.

Let's plot the histogram of the simulated data and compare it with the theoretical distribution.
