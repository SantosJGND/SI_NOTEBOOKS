## Continuous Distributions

#### Conceptual Understanding

Continuous distributions are used to model continuous random variables. Continuous random variables are random variables that can take on any value within a specified range (e.g. the height of an individual, the amount of time it takes to complete a task, etc.).

Think that the difference between different values of a continuous distribution can be infinitely small. For example, the difference between 1.5 and 1.6 is 0.1, but the difference between 1.5 and 1.51 is 0.01, and the difference between 1.5 and 1.501 is 0.001, and so on. 

As such, we consider that the probability of obtaining a specific value of a continuous distribution is 0. For example, the probability of someone existing a height of 1.500000001 meters is 0. Instead, we consider the probability of obtaining a value within a range of values. 


#### Probability Density Function (PDF)

Because of the nature of continuous distributions, we also "look" differently at their distritbutions. We use the probability density function (**PDF**) instead of the probability mass function (**PMF**). 

To understand the PDF we use the *area under the curve*, instead of the ~~height at specific points~~. The total area under the PDF curve is equal to 1. But the area under a specific point is 0.




### 3. The Normal Distribution

### Importance in Approximating Real-World Phenomena

The normal distribution is arguably the most important distribution in statistics. It is used to model a large number of real-world phenomena, such as the heights of individuals in a population, the systolic blood pressure of patients, test scores, and more. The normal distribution is also used in the fields of natural and social sciences to represent real-valued random variables whose distributions are not known.

### Bridging the Gap Between Theory and Real-World Phenomena

The normal distribution is a continuous probability distribution that is symmetric about the mean. The normal distribution is characterized by two parameters: the mean $\mu$ and the standard deviation $\sigma$. The normal distribution is defined as follows:

$$f(x) = \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}$$

where $\mu$ is the mean and $\sigma$ is the standard deviation.

#### Mean and Standard Deviation

In the case of the normal distribution, this distribution is already defined by the mean and standard deviation, so no need to calculate them. The mean $\mu$ is the center of the distribution, and the standard deviation $\sigma$ is the measure of the spread of the distribution.

#### Example

Let's look at a plot of the normal distribution for $\mu = 0$ and $\sigma = 1$. We will use the `norm` class of `scipy.stats` to generate the probability density function of the normal distribution for a set of parameters.

Notice the use of the `pdf` method of the `norm` class. Normal outcomes are continuous, so we use the `pdf` method instead of the `pmf` method.

```python
# plot normal distribution for mean = 0 and standard deviation = 1

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the mean and standard deviation of the normal distribution
mean = 0
std_dev = 1

# Define the variable X
X = -1.5

# Generate x values for the plot
x = np.linspace(-5, 5, 100)

# Calculate the PDF and CDF values
pdf = norm.pdf(x, mean, std_dev)
cdf = norm.cdf(x, mean, std_dev)

# Calculate the CDF value at X and round it to 3 decimal places
cdf_X = round(norm.cdf(X, mean, std_dev), 3)

# Create the plot
plt.plot(x, pdf, label='PDF')
plt.fill_between(x, pdf, where=(x <= X), color='red', alpha=0.5, label='Area under curve')

# Set the x-axis limit
plt.xlim(-5, 5)

# Add the CDF value to the title
plt.title(f'mu = {mean}, sigma = {std_dev}, CDF at X = {X} is {cdf_X}')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

Take a moment to understand the plot above (normal distribution for $\mu = 0$ and $\sigma = 1$). The x-axis represents the values of the random variable, and the y-axis represents the probability of obtaining that value. For example, the probability of obtaining a value between -1 and 1 is approximately 0.68.

What is the probability of obtaining a value between -2 and 2? What is the probability of obtaining a value between -3 and 3? What is the probability of obtaining a value between -4 and 4? What is the probability of obtaining a value between -5 and 5?

Let's use the `norm` class of `scipy.stats` to answer these questions.

```python
# use norm class to ask questions about the distribution

dist = norm(mean, std_dev)

# what is the probability of getting a value less than 0?
print(f"Probability of getting a value less than 0: {dist.cdf(0):.4f}")

# what is the probability of getting a value greater than 0?

print(f"Probability of getting a value greater than 0: {1 - dist.cdf(0):.4f}")

# what is the probability of getting a value between -1 and 1?
print(f"Probability of getting a value between -1 and 1: {dist.cdf(1) - dist.cdf(-1):.4f}")

```

### The Standard Normal Distribution

The standard normal distribution is a special case of the normal distribution. The standard normal distribution is a normal distribution with $\mu = 0$ and $\sigma = 1$. The standard normal distribution follows the same formula as the normal distribution:

$$f(x) = \frac{1}{\sqrt{2\pi}}e^{-\frac{1}{2}x^2}$$

but with $\mu = 0$ and $\sigma = 1$.

### Importance of the Standard Normal Distribution

The standard normal distribution draws its value from the fact that every normal distribution can be transformed into a standard normal distribution. This means that we can focus our understanding on the standard normal distribution and apply that understanding to any normal distribution.

### Standardizing a Normal Distribution

To transform a normal distribution into a standard normal distribution, we use the following formula:

$$z = \frac{x - \mu}{\sigma}$$

where $z$ is the standard normal random variable, $x$ is the normal random variable, $\mu$ is the mean of the normal distribution, and $\sigma$ is the standard deviation of the normal distribution.

#### Example



```python
## plot two normal distributions on two subplots side by side, using histograms.
## first, generate a set of 1000 random numbers from a normal distribution with mean =  20 and standard deviation = 5
## then, standardize the data by subtracting the mean and dividing by the standard deviation
## then, plot the data on a histogram with 20 bins

import numpy as np
import matplotlib.pyplot as plt


# Generate the data
data = np.random.normal(20, 5, 1000)

# Standardize the data
standardized_data = (data - np.mean(data)) / np.std(data)

# Create the plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.hist(data, bins=30)
ax1.set_title('Original Data')
ax2.hist(standardized_data, bins=30)
ax2.set_title('Standardized Data')
# larger ticks on both axis for readability
ax1.tick_params(axis='both', which='major', labelsize=12)
ax2.tick_params(axis='both', which='major', labelsize=12)

plt.show()

```

### 4. The Exponential Distribution

### Link to the Poisson Distribution


In the Poisson distribution, we were interested in the number of events that occurred in a given time interval. Imagine now that we are interested in the amount of time that passes between events. For example, the amount of time that passes between two customers entering a store, or the amount of time that passes between two earthquakes.

The exponential distribution is a continuous probability distribution that is used to model the amount of time that passes between events in a Poisson process. It characterizes the distribution of waiting time. 

### Bridging the Gap Between Theory and Real-World Phenomena

The exponential distribution is defined as follows:

$$f(x) = \lambda e^{-\lambda x}$$

where $\lambda$ is the rate parameter. The rate parameter is the average number of events per unit of time. The unit of time can be anything, such as a minute, an hour, a day, etc.

##### Mean and Standard Deviation

- The mean of the exponential distribution is equal to $\frac{1}{\lambda}$.
- The standard deviation of the exponential distribution is equal to $\frac{1}{\lambda}$.

#### Example

Let's look at a plot of the exponential distribution for $\lambda = 1$. We will use the `expon`.




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Set the lambda parameter for the exponential distribution
lmbda = 1.5

# Generate x values for the plots
x = np.linspace(0, 5, 100)

# Calculate the PDF and CDF values for the exponential distribution
pdf = expon.pdf(x, scale=1/lmbda)
cdf = expon.cdf(x, scale=1/lmbda)

# Create the figure and subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Plot the exponential distribution on the left subplot
axs[0].plot(x, pdf, label='PDF')
axs[0].fill_between(x, pdf, where=(x <= 1.5), color='red', alpha=0.5)
axs[0].set_title('Exponential Distribution')
axs[0].set_xlabel('x')
axs[0].set_ylabel('Probability Density')

# Plot the CDF of the exponential distribution on the right subplot
axs[1].plot(x, cdf, label='CDF')
axs[1].set_title('Cumulative Distribution Function')
axs[1].set_xlabel('x')
axs[1].set_ylabel('Cumulative Probability')

# Add legend for the area under the curve
axs[0].legend(loc='upper right')
axs[1].legend(loc='upper right')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the figure
plt.show()
```
