### Law of Large Numbers (LLN)

The Law of Large Numbers (LLN) is a fundamental theorem in probability and statistics that describes the behavior of sample averages as the sample size increases. It essentially states that as the sample size grows larger, the sample mean or average converges towards the population mean.

#### Types of Law of Large Numbers:

##### Weak Law of Large Numbers (WLLN):
- WLLN states that as the sample size increases, the sample mean approaches the population mean in probability.
- Formally, for a sequence of random variables \(X_1, X_2, \ldots, X_n\) with finite means, \( \bar{X}_n \rightarrow \mu\) in probability as \(n\) approaches infinity.

## Consequences for variance

One of the consequences of the LLN is that the variance of the sample mean decreases as the sample size increases. This is a consequence of the fact that the sample mean is an average of the sample values, and the variance of an average of independent random variables decreases as the number of variables increases.

But don't worry about the math, we'll see this in action in the next exercise. The important thing to remember is that the variance of the sample mean decreases as the sample size increases.
