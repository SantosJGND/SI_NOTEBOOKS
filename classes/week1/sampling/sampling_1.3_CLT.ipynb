# The Central Limit Theorem

So far we saw that:

- The sample mean converges to the population mean as the sample size increases.
- The variance of the sample mean decreases as the sample size increases.
- The LLN holds for any distribution.

But what if we are interested in the distribution of the sample mean itself? What can we say about the distribution of the sample mean?

Look back to the second plot of this lesson.

Go on. I'll wait.

The distribution of the sample mean is approximately normal. This is a consequence of the Central Limit Theorem (CLT). The CLT states that: 

- the sample mean of any independent random variable will be approximately normally distributed if the sample size is large enough.

It carries the same assumptions as the LLN, i.e., the samples are drawn independently and identically distributed (i.i.d.). Violation of these assumptions might affect the convergence of sample means. Additionally, the CLT is said to hold for "large enough" sample sizes (usually \(n \geq 30\)). However, it also applies to any distribution, like the LLN.

### Recap


1. Across repeated samples, $\bar{X}$ becomes a random variable. 
2. When we consider that samples are drawn from the same population, in the same manner and of the same size, then we can say that $\bar{X}$ is an *independent and identically distributed* **(i.i.d.)** random variable. Remember our slides. 
3. The distribution of $\bar{X}$ is normal, with mean $\mu$ and variance $\frac{\sigma^2}{n}$.

### Proof of the Distribution of $\bar{X}$

$$
X_1, X_2, \ldots, X_n \sim \text{Normal}(\mu, \sigma^2)
$$

Now, let's consider the random variable $\bar{X}$. Recall that $\bar{X}$ is the sample mean, and that it is defined as

$$
\bar{X} = \frac{X_1 + X_2 + \ldots + X_n}{n}
$$

Therefore, we can write

$$
\bar{X} = \frac{1}{n} \sum_{i=1}^n X_i
$$

Now, let's consider the expected value of $\bar{X}$. Recall that the expected value of a sum of random variables is equal to the sum of the expected values of the random variables. Therefore, we can write

$$
\begin{aligned}
E(\bar{X}) &= E\left(\frac{1}{n} \sum_{i=1}^n X_i\right) \\
&= \frac{1}{n} \sum_{i=1}^n E(X_i) \\
&= \frac{1}{n} \sum_{i=1}^n \mu \\
&= \frac{1}{n} \cdot n \cdot \mu \\
&= \mu
\end{aligned}
$$

### Practical Implications of CLT:

- **Statistical Inference:** The CLT underpins much of statistical inference. It allows us to make inferences about population parameters (such as the population mean) based on sample statistics, assuming normality.

- **Hypothesis Testing and Confidence Intervals:** It justifies the use of parametric tests and confidence intervals, assuming normality for estimators and allowing for the calculation of critical values.

- **Real-world Applications:** The CLT is essential in fields like quality control, finance, social sciences, and more, where understanding the distribution of sample means is crucial for decision-making.
