## The Dehli Temperature Dataset 

This data set contains daily temperata data in Dehli, India, between 1990 and 2021. 

The Data was originally Downloaded from the

- Weather data in Indian cities (1990-2021) [Kaggle](https://www.kaggle.com/datasets/vanvalkenberg/historicalweatherdataforindiancities) data set.
- A `season` column was added to the data set, based on the month of the year.


````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
filename= "data/Delhi_NCR_1990_2022_Safdarjung_seasons.csv"
dehli_df= pd.read_csv(filename)

dehli_df.head()
dehli_df["date"]= dehli_df["time"].apply(lambda x: datetime.strptime(x, "%d-%m-%Y"))
dehli_df.head()

````

````python
# describe the data
dehli_df.describe()
````

````python
# plot histogram of tavg

import matplotlib.pyplot as plt
import seaborn as sns


sns.set_style("darkgrid")
plt.figure(figsize=(12, 6))
plt.title("Average Temperature Distribution")
sns.histplot(data=dehli_df, x="tavg", bins=30)
plt.show()
````

````python
# plot histogram of tavg by season
# ignore future warning
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

sns.set_style("darkgrid")
plt.figure(figsize=(12, 6))
plt.title("Temperature Distribution")
sns.histplot(data=dehli_df, x="tavg", hue="season", bins=30)
plt.show()
````

### Subsetting the data

We will want to compare the average temperature for the month of June in 1998 against the average temperature for the month of June in 2012.  To do this, we will need to subset the data for the relevant years.

**subsetting by date**
Notice that in the import cell, we have made sure that the date column is in datetime format.  This will make it easier to subset the data by year and/or month.


```python
df_1998 = dehli_df[dehli_df.date.dt.year == 1998]
```

to subset for a given month, we can use the following code:

```python
df_june = dehli_df[dehli_df.date.dt.month == 6]
```

**subsetting by season**

The data set contains a `season` column with four seasons: winter, spring, summer and fall.  To subset the data for a given season, we can use the following code:

```python
df_summer = dehli_df[dehli_df.season == 'Summer']
```

**Multiple conditions**

We can also subset the data for multiple conditions.  For example, if we want to subset the data for the month of June in 1998, we can use the following code:

```python
df_june_1998 = dehli_df[(dehli_df.date.dt.year == 1998) & (dehli_df.date.dt.month == 6)]
```
